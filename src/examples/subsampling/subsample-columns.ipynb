## Subsampling columns of a table in iventure

__To create a new table with random subsample of columns from another table__:
```
usage .subsample_columns [options] <table> <new_table> <limit>

[options]
    --keep=[col1,col2,col3...]  List of columns that must be kept.
```

````python
%load_ext iventure.magics
````

````python
!rm -f resources/gapminder.subsample.bdb
%bayesdb resources/gapminder.subsample.bdb
````

````python
%bql CREATE TABLE gapminder FROM 'resources/gapminder.csv'
%bql .nullify gapminder ''
````

````python
%bql DROP TABLE IF EXISTS gapminder_subsample
%bql .subsample_columns --keep=country gapminder gapminder_subsample 10
````

````python
%bql SELECT * FROM gapminder_subsample LIMIT 15;
````
